# Introduction to Matplotlib
Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. It provides an object-oriented API for embedding plots into applications.

In this notebook, we'll cover the basics of Matplotlib, including creating simple plots, customizing them, and displaying multiple plots.

```python
# Importing Matplotlib
import matplotlib
import matplotlib.pyplot as plt

# Check the version of Matplotlib
print("Matplotlib version:", matplotlib.__version__)
```

## Creating a Simple Plot
Let's start by creating a basic line plot. We'll plot some simple data points to get familiar with the basic plotting functionality.

```python
# Create a simple line plot
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y, marker='o')
plt.title('Simple Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
print("Simple line plot created.")
```

## Customizing Plots
Matplotlib allows for extensive customization of plots, including line styles, markers, colors, and more. Let's explore some of these options.

```python
# Customized plot
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y, linestyle='--', color='r', marker='s', markersize=10)
plt.title('Customized Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
print("Customized plot created.")
```

## Displaying Multiple Plots
You can create multiple plots in a single figure using Matplotlib's `subplots` function. Let's create a figure with multiple subplots.

```python
# Multiple subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot in the first subplot
axs[0, 0].plot(x, y, 'tab:blue')
axs[0, 0].set_title('Line Plot')

# Scatter plot in the second subplot
axs[0, 1].scatter(x, y, color='tab:orange')
axs[0, 1].set_title('Scatter Plot')

# Bar plot in the third subplot
axs[1, 0].bar(x, y, color='tab:green')
axs[1, 0].set_title('Bar Plot')

# Histogram in the fourth subplot
axs[1, 1].hist(y, bins=5, color='tab:red')
axs[1, 1].set_title('Histogram')

plt.tight_layout()
plt.show()
print("Multiple plots created.")
```
