# Introduction to Seaborn
Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. It makes it easy to create complex plots with minimal code.

In this notebook, we'll cover the basics of Seaborn, including creating simple plots, customizing them, and using built-in datasets for visualization.

```python
# Importing Seaborn
import seaborn as sns
import matplotlib.pyplot as plt

# Check the version of Seaborn
print("Seaborn version:", sns.__version__)
```

## Creating Basic Plots
Seaborn provides several built-in functions to create various types of plots. We'll start by creating some basic plots using Seaborn.

```python
# Load an example dataset
df = sns.load_dataset('iris')

# Create a scatter plot
sns.scatterplot(x='sepal_length', y='sepal_width', data=df)
plt.title('Sepal Length vs Sepal Width')
plt.show()

# Create a histogram
sns.histplot(df['sepal_length'], kde=True)
plt.title('Sepal Length Distribution')
plt.show()

# Create a box plot
sns.boxplot(x='species', y='sepal_length', data=df)
plt.title('Sepal Length by Species')
plt.show()
print("Basic plots created.")
```

## Customizing Plots
Seaborn allows for customization of plots including colors, styles, and more. Let's explore how to customize the appearance of plots.

```python
# Customized scatter plot
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', style='species', palette='deep', data=df)
plt.title('Customized Scatter Plot')
plt.show()

# Customized box plot
sns.boxplot(x='species', y='sepal_length', palette='pastel', data=df)
plt.title('Customized Box Plot')
plt.show()
print("Customized plots created.")
```

## Using Built-in Datasets
Seaborn comes with several built-in datasets for practice. Let's explore one of these datasets to see how Seaborn can be used for data exploration and visualization.

```python
# Load another example dataset
df_tips = sns.load_dataset('tips')

# Create a pairplot
sns.pairplot(df_tips, hue='day')
plt.title('Pairplot of Tips Dataset')
plt.show()
print("Built-in dataset visualized.")
```
